# Remove Unnecessary Columns

Columns that do not contribute to prediction are removed to simplify the dataset.

```python
df1=df.drop(['url','address','phone','dish_liked','reviews_list','menu_item'],axis=1)
df1
```

```python
df1.drop_duplicates(inplace=True)
```

## Verify Missing Values

Check missing values after removing unnecessary columns.

```python
df1.isna().sum()
```

# Rename Columns

Rename long column names to make the dataset easier to work with.

```python
df1=df1.rename(columns={'approx_cost(for two people)':'cost', 'listed_in(type)':'type','listed_in(city)':'city'})
df1
```

## Explore Cost Column

Inspect the values in the restaurant cost column.

```python
df1['cost'].unique()
```

## Clean Cost Column

Remove commas and convert the cost column into numeric format.

```python
df1['cost']=df1['cost'].str.replace(',','')
df1['cost'].unique()
```

## Explore Rating Column

Inspect the unique rating values before cleaning.

```python
df1['rate'].unique()
```

## Clean Rating Values

Remove invalid rating values such as "NEW", "-" and "/5".

```python
#remove NEW  /5  -

df1['rate']=df1['rate'].str.replace('/5','')
df1['rate']=df1['rate'].str.replace('NEW','null')
df1['rate']=df1['rate'].str.replace('-','null')

df1['rate']=df1['rate'].str.replace('null','0')
df1['rate'].unique()
```

## Convert Ratings to Numeric

Convert the cleaned rating values into floating-point numbers.

```python
df1['rate']=df1['rate'].astype(float)
df1['cost']=df1['cost'].astype(float)
```

```python
df1.dtypes
```

## Handle Zero Ratings

Replace invalid zero ratings with missing values.

```python
df1.loc[df1.rate==0,'rate']=np.NaN
df1
```

```python
df1.isna().sum()
```

## Fill Missing Ratings

Replace missing ratings using the mean rating.

```python
df1['rate'].fillna(df1['rate'].mean(),inplace=True)
df1['rate'].isna().sum()
```

```python
df1.isna().sum()
```

## Remove Remaining Missing Values

Drop rows that still contain missing values after preprocessing.

```python
df1.dropna(inplace=True)
df1.isna().sum()
```

## Encoding Binary Categorical Features

Convert binary categorical variables into numerical values to make them suitable for machine learning algorithms, which require numerical input for model training.

```python
df1['online_order']=df1['online_order'].map({'Yes':1,'No':0})
df1['book_table']=df1['book_table'].map({'Yes':1,'No':0})
df1
```

### Observation

The categorical values in the **online_order** and **book_table** features have been successfully converted into binary numerical values (**Yes = 1**, **No = 0**), making the dataset ready for further preprocessing and model development.

## Encoding Categorical Features

Transform categorical variables into numerical representations so they can be processed effectively by machine learning algorithms during model training.

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1.location = le.fit_transform(df1.location)
df1.rest_type = le.fit_transform(df1.rest_type)
df1.cuisines = le.fit_transform(df1.cuisines)
df1.type = le.fit_transform(df1.type)
df1
```

### Observation

The selected categorical features have been successfully encoded into numerical values, making the dataset fully compatible with machine learning models while preserving the unique categories within each feature.

## Selecting Input Features

Prepare the feature set by selecting the independent variables that will be used to train the machine learning models for restaurant rating prediction.

```python
x=df2.iloc[:,[0,1,3,4,5,6,7,8]]
x
```

## Defining the Target Variable

Select the target variable that the machine learning models will learn to predict during the training process.

```python
y=df2['rate']
y
```

## Splitting the Dataset

Divide the dataset into training and testing sets to evaluate the model's performance on unseen data and ensure its ability to generalize effectively.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

## Linear Regression Model

Train a Linear Regression model to predict restaurant ratings based on the selected features. This model serves as a baseline for evaluating the performance of more advanced machine learning algorithms.

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

## Comparing Actual and Predicted Ratings

Create a comparison table to evaluate how closely the model's predicted ratings match the actual restaurant ratings.

```python
df3=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred})
df3
```

## Model Evaluation Using R² Score

Evaluate the performance of the Linear Regression model by measuring how well it explains the variation in restaurant ratings.

```python
from sklearn.metrics import r2_score
r2_score1=r2_score(y_test,y_pred)
r2_score1
```

### Interpretation

The Linear Regression model achieved an **R² score of 0.273**, indicating that it explains approximately **27.3%** of the variation in restaurant ratings.

### Key Observation

- The model captures only a small portion of the relationship between the input features and restaurant ratings.
- A significant amount of variation remains unexplained, suggesting that the relationship between the features and ratings is not purely linear.

### Business Insight

Restaurant ratings are influenced by multiple complex factors, including customer preferences, service quality, food quality, and brand reputation. A simple Linear Regression model may not fully capture these non-linear relationships.

### Recommendation

While Linear Regression provides a useful baseline model, more advanced algorithms such as **Decision Tree** or **Random Forest** should be explored to improve prediction accuracy and better capture complex patterns in the data.

## Decision Tree Regressor

Train a Decision Tree Regressor to predict restaurant ratings. Unlike Linear Regression, this model can capture complex, non-linear relationships between the features and the target variable, making it suitable for structured datasets.

```python
from sklearn.tree import DecisionTreeRegressor
tree=DecisionTreeRegressor()
tree.fit(x_train,y_train)
y_pred1=tree.predict(x_test)
y_pred1
```

### Prediction Results

A comparison of the actual and predicted restaurant ratings is created to visually assess how closely the Decision Tree model's predictions align with the true values.

```python
df4=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred1})
df4
```

```python
from sklearn.metrics import r2_score
r2_score2=r2_score(y_test,y_pred1)
r2_score2
```

### Model Evaluation

The Decision Tree Regressor achieved an **R² score of 0.872**, indicating that it explains approximately **87.2%** of the variation in restaurant ratings.

### Key Observations

- The model demonstrates a substantial improvement over the Linear Regression model.
- The high R² score indicates that the Decision Tree effectively captures the underlying patterns in the dataset.
- The model is well-suited for handling the non-linear relationships between restaurant features and ratings.

### Business Insight

The strong predictive performance suggests that factors such as customer votes, cost, location, restaurant type, and service availability interact in complex ways to influence restaurant ratings. Decision Tree successfully captures these interactions, resulting in more accurate predictions.

### Recommendation

The Decision Tree Regressor is a strong candidate for restaurant rating prediction due to its high accuracy. However, since Decision Trees are prone to overfitting, ensemble methods such as **Random Forest** or **XGBoost** should also be evaluated to determine whether they provide even better generalization on unseen data.

## Random Forest Regressor

Train a Random Forest Regressor to predict restaurant ratings. This ensemble learning algorithm combines multiple decision trees to improve prediction accuracy and reduce the risk of overfitting.

```python
from sklearn.ensemble import RandomForestRegressor
forest=RandomForestRegressor()
forest.fit(x_train,y_train)
y_pred2=forest.predict(x_test)
y_pred2
```

### Prediction Results

A comparison of the actual and predicted restaurant ratings is created to evaluate how accurately the Random Forest model predicts unseen data.

```python
df5=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred2})
df5
```

```python
from sklearn.metrics import r2_score
r2_score3=r2_score(y_test,y_pred2)
r2_score3
```

### Random Forest Evaluation

The Random Forest Regressor achieved an **R² score of 0.913**, indicating that it explains approximately **91.3%** of the variation in restaurant ratings.

### Key Observations

- The Random Forest model outperformed both the Linear Regression and Decision Tree models.
- The high R² score demonstrates excellent predictive performance and strong generalization capability.
- Combining multiple decision trees enables the model to capture complex patterns while reducing overfitting.

### Business Insight

The superior performance of the Random Forest model suggests that restaurant ratings are influenced by multiple interacting factors rather than simple linear relationships. By leveraging ensemble learning, the model provides highly reliable rating predictions, making it suitable for real-world applications such as restaurant recommendation systems and customer decision support.

### Recommendation

Based on its high predictive accuracy, the **Random Forest Regressor** is the preferred model for this project. It can be effectively used to predict restaurant ratings and support business decisions, such as identifying high-performing restaurants, improving recommendation systems, and analyzing the impact of key restaurant features on customer ratings.

## Model Performance Comparison

Compare the performance of all regression models using their **R² scores** to identify the model that provides the most accurate restaurant rating predictions.

# Limitations

While the project achieved strong predictive performance, it has a few limitations:

- The dataset contains only the available restaurant attributes and does not include factors such as food quality, customer reviews, service quality, ambiance, or delivery experience, which can significantly influence restaurant ratings.
- Categorical features were encoded using **Label Encoding**, which assigns numerical values without preserving any meaningful relationship between categories.
- The model was trained on historical data and may not accurately predict ratings for newly established restaurants or changing market trends.
- Hyperparameter tuning was not performed, so there is potential to further improve the model's performance.
- The project evaluated only three regression algorithms. Other advanced models, such as **XGBoost**, **LightGBM**, and **CatBoost**, may achieve even better predictive accuracy.
- The dataset represents restaurants from a specific region, so the model's performance may not generalize well to restaurants in other cities or countries without retraining.
- Although the Random Forest model achieved high accuracy, ensemble models are generally less interpretable than simpler models such as Linear Regression.
